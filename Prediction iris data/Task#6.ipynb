<u><b>Context : <b></u>
     
<p>The Iris flower data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems. It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species. The data set consists of 50 samples from each of three species of Iris (Iris Setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

This dataset became a typical test case for many statistical classification techniques in machine learning such as support vector machines

<u><b>Content : </b></u>
    
The dataset contains a set of 150 records under 5 attributes - Petal Length, Petal Width, Sepal Length, Sepal width and Class(Species).

<font size = 5px color = 'red'><b>Decision Tree</center></font>
 

#### Split data into Features and Target

```python
Features = df_iris.drop(columns='Species')
Target = df_iris.Species
```

#### Kfold

```python
kf = KFold(n_splits= 5, shuffle=True, random_state=1)
```

```python
# Creating Estimator Object
decision_tree = DecisionTreeClassifier(random_state=1)

# Fit Model 
decision_tree.fit(Features,Target)

# Model Score:
decision_tree_Accuracy = model_selection.cross_val_score(decision_tree,Features,Target,cv=kf,scoring='accuracy').mean()*100
print("\nAccuracy  : ", round(decision_tree_Accuracy,3),"%\n" )

# Classification Report
print("Classfication_Report : \n\n", metrics.classification_report(Target,model_selection.cross_val_predict(decision_tree,Features,Target)))

# Cross_tab
print(pd.crosstab(Target,model_selection.cross_val_predict(decision_tree,Features,Target)))
```

```python
# Creating Estimator Object
xgboost = XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.85, eval_metric='error',
              gamma=1, gpu_id=-1, importance_type='gain',
              interaction_constraints='', learning_rate=0.5, max_delta_step=0,
              max_depth=2, min_child_weight=0.01,
              monotone_constraints='()', n_estimators=750, n_jobs=4,
              num_parallel_tree=1, objective='multi:softprob', random_state=0,
              reg_alpha=0.5, reg_lambda=15, sampling_method='uniform',
              scale_pos_weight=None, subsample=0.55, tree_method='exact',
              validate_parameters=1, verbosity=None)

# Fit Model 
xgboost.fit(Features,Target)

# Model Score:
xgboost_Accuracy = model_selection.cross_val_score(xgboost,Features,Target,cv=kf,scoring='accuracy').mean()*100
print("\nAccuracy  : ", round(xgboost_Accuracy,3),"%\n" )

# Classification Report
print("Classfication_Report : \n\n", metrics.classification_report(Target,model_selection.cross_val_predict(xgboost,Features,Target)))

# Cross_tab
print(pd.crosstab(Target,model_selection.cross_val_predict(xgboost,Features,Target)))
```
